```python
# Imports
from pycoingecko import CoinGeckoAPI
import pandas as pd
import datetime
import time
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from pathlib import Path
import seaborn as sns
%matplotlib inline
# initilize client
client = CoinGeckoAPI()
client.ping()
```

```python
def main():
    # Helper Functions
    def unix_time(year, month, day, hour, second):
        date_time = datetime.datetime(year, month, day, hour, second)
        return time.mktime(date_time.timetuple())

    def human_time(unix_time):
        return datetime.datetime.fromtimestamp(unix_time)
    
    start = unix_time(2018, 1, 1, 0, 0)
    end = unix_time(2023, 1, 1, 0, 0)
    
    # Getting input for specific coins/ dates
    coin1 = input("What is the first coin you would like to compare")
    coin2 = input("What is the second coin you would like to compare")
    coin3 = input("What is the third coin you would like to compare")
    coin4 = input("What is the fourth coin you would like to compare")
    coin5  = input("What is the fith coin you would like to compare")
    coin1 = coin1.lower()
    coin2 = coin2.lower()
    coin3 = coin3.lower()
    coin4 = coin4.lower()
    coin5 = coin5.lower()
    
    
    # Reading in specified dates from API
    user_coin1 = client.get_coin_market_chart_range_by_id(
    id = coin1,
    vs_currency = 'usd',
    from_timestamp=start,
    to_timestamp=end)
    
    user_coin2 = client.get_coin_market_chart_range_by_id(
    id = coin2,
    vs_currency = 'usd',
    from_timestamp=start,
    to_timestamp=end)
    
    user_coin3 = client.get_coin_market_chart_range_by_id(
    id = coin3,
    vs_currency = 'usd',
    from_timestamp=start,
    to_timestamp=end)
    
    user_coin4 = client.get_coin_market_chart_range_by_id(
    id = coin4,
    vs_currency = 'usd',
    from_timestamp=start,
    to_timestamp=end)
    
    user_coin5 = client.get_coin_market_chart_range_by_id(
    id = coin5,
    vs_currency = 'usd',
    from_timestamp=start,
    to_timestamp=end)
    
    # Making dataframes from API
    coin1_df = pd.DataFrame.from_dict(user_coin1)
    coin2_df= pd.DataFrame.from_dict(user_coin2)
    coin3_df = pd.DataFrame.from_dict(user_coin3)
    coin4_df = pd.DataFrame.from_dict(user_coin4)
    coin5_df = pd.DataFrame.from_dict(user_coin5)
    
    # Cleaning Coin1Data
    coin1_df['date'] = coin1_df['prices'][0][0]
    coin1_df[['time', 'price']] = coin1_df['prices'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin1_df[['time2', 'market_cap']] = coin1_df['market_caps'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin1_df[['time3', 'total_volume']] = coin1_df['total_volumes'].apply(lambda x: pd.Series([x[0], x[1]]))
    
    coin1_df.drop('prices', axis=1, inplace=True)
    coin1_df.drop('market_caps', axis=1, inplace=True)
    coin1_df.drop('total_volumes', axis=1, inplace=True)
    coin1_df.drop('date', axis=1, inplace=True)
    coin1_df.drop('time2', axis=1, inplace=True)
    coin1_df.drop('time3', axis=1, inplace=True)
    coin1_df['time'] = pd.to_datetime(coin1_df['time'], unit='ms')
    
    #Cleaning Coin2 Data
    coin2_df['date'] = coin2_df['prices'][0][0]
    coin2_df[['time', 'price']] = coin2_df['prices'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin2_df[['time2', 'market_cap']] = coin2_df['market_caps'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin2_df[['time3', 'total_volume']] = coin2_df['total_volumes'].apply(lambda x: pd.Series([x[0], x[1]]))
    
    coin2_df.drop('prices', axis=1, inplace=True)
    coin2_df.drop('market_caps', axis=1, inplace=True)
    coin2_df.drop('total_volumes', axis=1, inplace=True)
    coin2_df.drop('date', axis=1, inplace=True)
    coin2_df.drop('time2', axis=1, inplace=True)
    coin2_df.drop('time3', axis=1, inplace=True)
    coin2_df['time'] = pd.to_datetime(coin2_df['time'], unit='ms')
    
    # Cleaning coin 3 Data
    coin3_df['date'] = coin3_df['prices'][0][0]
    coin3_df[['time', 'price']] = coin3_df['prices'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin3_df[['time2', 'market_cap']] = coin3_df['market_caps'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin3_df[['time3', 'total_volume']] = coin3_df['total_volumes'].apply(lambda x: pd.Series([x[0], x[1]]))
    
    coin3_df.drop('prices', axis=1, inplace=True)
    coin3_df.drop('market_caps', axis=1, inplace=True)
    coin3_df.drop('total_volumes', axis=1, inplace=True)
    coin3_df.drop('date', axis=1, inplace=True)
    coin3_df.drop('time2', axis=1, inplace=True)
    coin3_df.drop('time3', axis=1, inplace=True)
    coin3_df['time'] = pd.to_datetime(coin3_df['time'], unit='ms')
    
    # Cleaning coin 4 Data
    coin4_df['date'] = coin4_df['prices'][0][0]
    coin4_df[['time', 'price']] = coin4_df['prices'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin4_df[['time2', 'market_cap']] = coin4_df['market_caps'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin4_df[['time3', 'total_volume']] = coin4_df['total_volumes'].apply(lambda x: pd.Series([x[0], x[1]]))
    
    coin4_df.drop('prices', axis=1, inplace=True)
    coin4_df.drop('market_caps', axis=1, inplace=True)
    coin4_df.drop('total_volumes', axis=1, inplace=True)
    coin4_df.drop('date', axis=1, inplace=True)
    coin4_df.drop('time2', axis=1, inplace=True)
    coin4_df.drop('time3', axis=1, inplace=True)
    coin4_df
    coin4_df['time'] = pd.to_datetime(coin4_df['time'], unit='ms')
    
    # Cleaning coin 5 Data
    coin5_df['date'] = coin5_df['prices'][0][0]
    coin5_df[['time', 'price']] = coin5_df['prices'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin5_df[['time2', 'market_cap']] = coin5_df['market_caps'].apply(lambda x: pd.Series([x[0], x[1]]))
    coin5_df[['time3', 'total_volume']] = coin5_df['total_volumes'].apply(lambda x: pd.Series([x[0], x[1]]))
    
    coin5_df.drop('prices', axis=1, inplace=True)
    coin5_df.drop('market_caps', axis=1, inplace=True)
    coin5_df.drop('total_volumes', axis=1, inplace=True)
    coin5_df.drop('date', axis=1, inplace=True)
    coin5_df.drop('time2', axis=1, inplace=True)
    coin5_df.drop('time3', axis=1, inplace=True)
    coin5_df['time'] = pd.to_datetime(coin5_df['time'], unit='ms')
    
    # cleaning continued
    coin1_df = coin1_df.set_index('time')
    coin2_df = coin2_df.set_index('time')
    coin3_df = coin3_df.set_index('time')
    coin4_df = coin4_df.set_index('time')
    coin5_df = coin5_df.set_index('time')
    
    coin1_cleaned = coin1_df.drop(columns=['market_cap', 'total_volume'])
    coin2_cleaned = coin2_df.drop(columns=['market_cap', 'total_volume'])
    coin3_cleaned = coin3_df.drop(columns=['market_cap', 'total_volume'])
    coin4_cleaned = coin4_df.drop(columns=['market_cap', 'total_volume'])
    coin5_cleaned = coin5_df.drop(columns=['market_cap', 'total_volume'])
    # combined data 
    combined_data = pd.concat([coin1_cleaned, coin2_cleaned, coin3_cleaned, coin4_cleaned, coin5_cleaned], axis= 1, join="inner")
    
    #clean new data and find pct change
    combined_data = combined_data.pct_change().dropna()
    
    # find cumulative returns and plot 
    #cumulative_returns = (1 + combined_data).cumprod() 
    #cum_plot = cumulative_returns.plot(figsize=(10,5))
    
    # Sharpe Ratios
    sharpe_ratios = (combined_data.mean()*252) / (combined_data.std() * np.sqrt(252))
    sharpe_plot = sharpe_ratios.plot(kind = "bar")
    
    print(sharpe_plot)
```

```python
main()
```
